```python
breed1_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 307)), columns = ["Breed1_"+str(i) for i in range(1,308)])
breed2_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 307)), columns = ["Breed2_"+str(i) for i in range(1,308)])
gender_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 3)), columns = ["Male", "Female", "Group"])
color1_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 7)), columns = ["Color1_"+str(i) for i in range(1,8)])
color2_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 7)), columns = ["Color2_"+str(i) for i in range(1,8)])
maturity_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 4)), columns = ["Maturity_S", "Maturity_M", "Maturity_L", "Maturity_XL"])
fur_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 3)), columns = ["FurLength_S", "FurLength_M", "FurLength_L"]) 
vaccinated_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 2)), columns = ["Vaccinated_Y", "Vaccinated_N"])
dewormed_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 2)), columns = ["Dewormed_Y", "Dewormed_N"])
sterilized_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 2)), columns = ["Sterilized_Y", "Sterilized_N"])
health_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 3)), columns = ["Healthy", "MinorInjury", "SeriousInjury"])
state_onehot = pd.DataFrame(np.zeros(shape=(data.shape[0], 15)), columns = ["State_"+str(i) for i in [41336, 41325, 41367, 41401, 41415, 41324, 41332, 41335, 41330, 41380, 41327, 41345, 41342, 41326, 41361]])
```

```python
for i in range(breed1.shape[0]):
    if breed1[i] != 0:
        breed1_onehot["Breed1_"+str(breed1[i])][i] = 1
for i in range(breed2.shape[0]):
    if breed2[i] != 0:
        breed2_onehot["Breed2_"+str(breed2[i])][i] = 1
for i in range(gender.shape[0]):
    if gender[i] == 1:
        gender_onehot["Male"][i] = 1
    elif gender[i] == 2:
        gender_onehot["Female"][i] = 1
    elif gender[i] == 3:
        gender_onehot["Group"][i] = 1
for i in range(color1.shape[0]):
    if color1[i] != 0:
        color1_onehot["Color1_"+str(color1[i])][i] = 1
for i in range(color2.shape[0]):
    if color2[i] != 0:
        color2_onehot["Color2_"+str(color2[i])][i] = 1
for i in range(maturity.shape[0]):
    if maturity[i] == 1:
        maturity_onehot["Maturity_S"][i] = 1
    elif maturity[i] == 2:
        maturity_onehot["Maturity_M"][i] = 1
    elif maturity[i] == 3:
        maturity_onehot["Maturity_L"][i] = 1
    elif maturity[i] == 4:
        maturity_onehot["Maturity_XL"][i] = 1
for i in range(fur.shape[0]):
    if fur[i] == 1:
        fur_onehot["FurLength_S"][i] = 1
    elif fur[i] == 2:
        fur_onehot["FurLength_M"][i] = 1
    elif fur[i] == 3:
        fur_onehot["FurLength_L"][i] = 1
for i in range(vaccinated.shape[0]):
    if vaccinated[i] == 1:
        vaccinated_onehot["Vaccinated_Y"][i] = 1
    elif vaccinated[i] == 2:
        vaccinated_onehot["Vaccinated_N"][i] = 1
for i in range(dewormed.shape[0]):
    if dewormed[i] == 1:
        dewormed_onehot["Dewormed_Y"][i] = 1
    elif dewormed[i] == 2:
        dewormed_onehot["Dewormed_N"][i] = 1
for i in range(sterilized.shape[0]):
    if sterilized[i] == 1:
        sterilized_onehot["Sterilized_Y"][i] = 1
    elif sterilized[i] == 2:
        sterilized_onehot["Sterilized_N"][i] = 1
for i in range(health.shape[0]):
    if health[i] == 1:
        health_onehot["Healthy"][i] = 1
    elif health[i] == 2:
        health_onehot["MinorInjury"][i] = 1
    elif health[i] == 3:
        health_onehot["SeriousInjury"][i] = 1
for i in range(state.shape[0]):
    state_onehot["State_"+str(state[i])][i] = 1
```

```python
frames = [data, breed1_onehot, breed2_onehot, gender_onehot, color1_onehot, color2_onehot, maturity_onehot, fur_onehot, vaccinated_onehot, dewormed_onehot, sterilized_onehot, health_onehot, state_onehot]
df_categorical = pd.concat(frames, axis=1)
```
