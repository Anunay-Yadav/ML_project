```python
#keras model for late fusion
from tensorflow.keras import regularizers
image_input = Input(shape=(image_train_feat.shape[1],))
txt_input = Input(shape=(txt_train_feat.shape[1],))
csv_input = Input(shape=(csv_x.shape[1],))

#late fusion
h1_img = Dense(128,activation = "relu")(image_input)
h2_img = Dense(64,activation = "relu")(h1_img)
h2_img = Dropout(0.2)(h2_img)
h2_img = Dense(5,activation = "softmax")(h2_img)

h1_txt = Dense(256,activation = "relu")(txt_input)
h2_txt = Dense(64,activation = "relu")(h1_txt)
h2_txt = Dropout(0.2)(h2_txt)
h2_txt = Dense(5,activation = "softmax")(h2_txt)

h1_csv = Dense(128,activation = "relu")(csv_input)
h2_csv = Dense(64,activation = "relu")(h1_csv)
h2_csv = Dropout(0.2)(h2_csv)
h2_csv = Dense(5,activation = "softmax")(h2_csv)

z = Dense(8,activation = "sigmoid")(concatenate([h2_csv,h2_img,h2_txt]))
output = Dense(5,activation = "softmax")(z)


model_late = Model(inputs=[image_input,txt_input,csv_input], outputs=[output])

```

```python
# fitting modelon dataset
from keras import backend as K
from sklearn.model_selection import StratifiedShuffleSplit
from keras.optimizers import Adam
def get_f1(y_true, y_pred): #taken from old keras source code
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val


lr = 0.001
model_late.compile(optimizer=Adam(lr=lr),
              loss='categorical_crossentropy',
              metrics=['accuracy',get_f1])
model_late.fit([X_train_img,X_train_txt,X_train_csv], y_train, epochs=200, batch_size=128, validation_data=([X_test_img,X_test_txt,X_test_csv], y_test), callbacks=[NeptuneMonitor()])
```
